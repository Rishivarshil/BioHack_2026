```python
import pandas as pd

# Paths
csv_path = "organ_embeddings.csv"     # your input CSV
tsv_path = "tsv_embeddings.tsv"     # output TSV

# Load CSV
df = pd.read_csv(csv_path)

# Select only embedding columns (dim_0, dim_1, ..., dim_511)
embedding_cols = [col for col in df.columns if col.startswith("dim_")]

# Extract embeddings
embeddings = df[embedding_cols]

# Save as TSV without header or index
embeddings.to_csv(
    tsv_path,
    sep="\t",
    header=False,
    index=False,
    float_format="%.6f"
)

print(f"Saved {embeddings.shape[0]} vectors of dimension {embeddings.shape[1]} to {tsv_path}")
```
